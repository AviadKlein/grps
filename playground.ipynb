```python
from grps import GRPS
import networkx as nx
import matplotlib.pyplot as plt
```

```python
game = GRPS.the_office_version()
```

```python
game.play('Andy', 'Jim')
```

```python
g = game.get_nx()
```

```python
pos = nx.circular_layout(g)
```

```python
game.get_nx_edge_labels()
```

```python
plt.figure(figsize=(10,10))
pos = nx.shell_layout(g)
nx.draw(
    g, pos, edge_color='black', width=1, linewidths=1, font_size=15,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in g.nodes()}
)
nx.draw_networkx_edge_labels(
    g, pos,
    edge_labels=game.get_nx_edge_labels(),
    font_color='red', font_size=15
)
plt.axis('off')
plt.show()
```
